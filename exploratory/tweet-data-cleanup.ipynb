```python
import pandas as pd

tweets_df = pd.concat([pd.read_json('../dataset/tweet-processed.json', orient='index'), pd.read_excel('../dataset/tweet-label.xlsx').drop(['Unnamed: 0', 'raw_text', 'is_spam'], axis=1)], axis=1)
tweets_df.head()
```

```python
tweets_df.dropna(inplace=True)
```

```python
emotion_translate = {
    's': 'sad',
    'h': 'happy',
    'o': 'hopeful',
    'a': 'angry',
    'd': 'disgust',
    'n': 'neutral',
    'p': 'surprise',
    'w': 'funny',
    'f': 'fear'
}

tweets_df['emotion'] = tweets_df['emotion'].map(emotion_translate)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

plt.figure(figsize=(16,9))
sns.histplot(tweets_df, x='emotion', hue='emotion')
plt.show()
```
